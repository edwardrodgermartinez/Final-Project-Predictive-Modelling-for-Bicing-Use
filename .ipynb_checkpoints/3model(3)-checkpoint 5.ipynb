```python
#ARIMA
```

```python
train_size = int(len(df) * 0.9)
train, test = df.iloc[:train_size], df.iloc[train_size:]
```

```python
model = ARIMA(train['bikesInUsage'], order=(2, 0, 0))  # Specify appropriate values for p, d, q
result = model.fit()
```

```python
predictions = result.predict(start=len(train), end=len(train) + len(test) - 1, typ='levels')
```

```python
mse = mean_squared_error(test['bikesInUsage'], predictions)
rmse = np.sqrt(mse)
r2 = metrics.r2_score(test['bikesInUsage'], predictions)
```

```python
#SARIMA
```

```python
train_size = int(len(df) * 0.9)
train, test = df.iloc[:train_size], df.iloc[train_size:]
```

```python
order = (2, 0, 0)  # Specify appropriate values for p, d, q
seasonal_order = (2, 0, 0, 8)  # Specify appropriate values for P, D, Q, S

# Fit SARIMA model
sarima_model = sm.tsa.SARIMAX(train['bikesInUsage'], order=order, seasonal_order=seasonal_order)
sarima_result = sarima_model.fit()
```

```python
predictions = sarima_result.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)
```

```python
mse = mean_squared_error(test['bikesInUsage'], predictions)
rmse = np.sqrt(mse)
r2 = metrics.r2_score(test['bikesInUsage'], predictions)
```

```python
train_size = int(len(df) * 0.9)
train, test = df.iloc[:train_size], df.iloc[train_size:]

# Prepare data for Prophet
df_prophet_train = train.reset_index().rename(columns={'dateTime': 'ds', 'bikesInUsage': 'y'})

# Create and fit the Prophet model
model = Prophet()
model.fit(df_prophet_train)

# Make future DataFrame for predictions
future = model.make_future_dataframe(periods=len(test))

# Generate predictions
forecast = model.predict(future)
```

```python
test_predictions = forecast.iloc[-len(test):][['ds', 'yhat']]
test_predictions.set_index('ds', inplace=True)
```

```python
mse = mean_squared_error(test['bikesInUsage'], test_predictions['yhat'])
rmse = np.sqrt(mse)
r2 = metrics.r2_score(test['bikesInUsage'], test_predictions['yhat'])
```

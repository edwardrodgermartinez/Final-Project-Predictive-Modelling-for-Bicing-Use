```python
#Now going to check for stationarity
```

```python
res = adfuller(bicing_for_AR_model["bikesInUsage"])
res
```

```python
p_value = res[1]
```

```python
def dickey_rejects (bicing_for_AR_model, alpha = 0.05):
    
    p_value = adfuller(bicing_for_AR_model)[0]
    
    h0 = "The series is NOT stationary"
    h1 = "The series IS stationary"
    
    if p_value < alpha:
        return f"We reject H0 and accept H1: {h1}"
    
    elif p_value > alpha:
        return f"We don't have enough info to reject H0, so we keep H0: {h0}"
```

```python
dickey_rejects(bicing_for_AR_model["bikesInUsage"])
```

```python
#Amazing. 
#So we use AR and it is stationary
```

```python
from statsmodels.tsa.api import acf, pacf
from statsmodels.tsa.stattools import arma_order_select_ic

# Determine lag order using AIC
order_aic = arma_order_select_ic(bicing_for_AR_model['bikesInUsage'], ic='aic', trend='c')
print(f'Best lag order based on AIC: {order_aic["aic_min_order"]}')
```
